```python
import pandas as pd
```

```python
cases = pd.read_csv("Number_of_reported_cases.csv")
```

```python
cases
```

```python
cases.info()
```

```python
#convert col "Number of reported cases of cholera" to int64

def transformar(dato):
    num = int(dato)            
    return num
```

```python
# substitute spaces blank

cases['Number of reported cases of cholera'] = cases['Number of reported cases of cholera'].map(lambda x : x.replace(' ', ''))
cases["Number of reported cases of cholera"] = cases["Number of reported cases of cholera"].map(transformar)

```

```python
cases.info()
```

```python
cases = cases.dropna(axis = 1)
cases
```

```python
cases = cases.set_index("Country")
```

Cases by country

```python
cases_by_country = cases.groupby('Country')["Number of reported cases of cholera"].sum()
cases_by_country
```

```python
#transform it to dataframe

df_by_country = pd.DataFrame(cases_by_country)
df_by_country
```

Visualization

```python
import plotly.graph_objs as go

from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
```

```python
cases.reset_index(inplace = True)
cases
```

```python
df = cases.groupby(['Country','Year']).sum().reset_index()
df
```

```python
trace = go.Scattergl(
    x = df.Year,
    y = df.Country,
    mode = 'markers',
    text = df['Number of reported cases of cholera'],
    name = "casos",
    hoverinfo = "name+x+text"
)

data = [trace]

layout = dict(
    title = dict( text = "Cases and year relation" ),
    hovermode = "closest",
    hoverdistance = 10,
    
    # Changing labs to B/N
    # new font style
    hoverlabel = dict(
        bgcolor = "white",
        bordercolor = "black",
        font = dict(
            color = "black",
            family = "Georgia",
            size = 16
        )
    ),
    
    xaxis = dict(
        title = dict( text = "Year" ),
        tickvals = list(range(1910,2021,10)),
    ),
    yaxis = dict(
        title = dict( text = "Country" )
        #tickvals = list(range(0,11)),
    )
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
trace = go.Scatter(
    x = df.Year,
    y = df['Number of reported cases of cholera'],
    
    #text = df['Number of reported cases of cholera'],
    name = "casos",
    #hoverinfo = "name+x+text",
    mode = 'markers',
    text = df.Country,
        marker = dict(
            size = df['Number of reported cases of cholera'],
                sizemode="area",
            color = "mediumvioletred",
            sizeref = 100
            )
)

data = [trace]

layout = dict(
    title = dict( text = "Number of reported cases of cholera by year and country" ),
    hovermode = "closest",
    hoverdistance = 10,
    
    # Changing labs to B/N
    # new font style
    hoverlabel = dict(
        bgcolor = "white",
        bordercolor = "black",
        font = dict(
            color = "black",
            family = "Georgia",
            size = 16
        )
    ),
    
    xaxis = dict(
        title = dict( text = "Year" ),
        tickvals = list(range(1910,2021,10)),
    ),
    yaxis = dict(
        title = dict( text = "Cases" )
        #tickvals = list(range(0,11)),
    )
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
fig.write_html("Cholera.html")
```

```python
import plotly.express as px
fig = px.scatter(df, x= "Year",
    y = df['Number of reported cases of cholera'], 
    color='Country',
    color_discrete_sequence=px.colors.qualitative.Antique,
    size = df['Number of reported cases of cholera'],
    log_x=True, size_max=60,
    )
fig.update_layout(width=700,
    hovermode="closest",
    showlegend=False,
    title_text="Number of reported cases of cholera by year and country") 
fig.update_layout({'plot_bgcolor': 'rgba(0, 0, 0, 0)',
    'paper_bgcolor': 'rgba(0, 0, 0, 0)'})


fig.show()
with open('Cholera2.html', 'a') as f:
    f.write(fig.to_html(full_html=False, include_plotlyjs='cdn'))
```
